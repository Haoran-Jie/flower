```python
from flwr_datasets import FederatedDataset
from flwr_datasets.partitioner import GroupedNaturalIdPartitioner
from flwr_datasets.visualization import plot_label_distributions
import matplotlib.pyplot as plt
```

```python
partitioner = GroupedNaturalIdPartitioner(
    partition_by="speaker_id", group_size=30
)
fds = FederatedDataset(
    dataset="speech_commands",
    subset="v0.02",
    partitioners={"train": partitioner},
)
partition = fds.load_partition(0)
```

```python
fig, axis = plt.subplots(figsize=(16, 6))
fig, ax, df = plot_label_distributions(
    partitioner,
    axis=axis,
    label_name="label",
    plot_type="bar",
    size_unit="absolute",
    partition_id_axis="x",
    legend=True,
    verbose_labels=True,
    title="Per Partition Labels Distribution",
    legend_kwargs={"ncols": 2, "bbox_to_anchor": (1.05, 0.5)},    
)
```

```python
fig.savefig("whisper_flower_data.png", format="png", bbox_inches="tight")
```
